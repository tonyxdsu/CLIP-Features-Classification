```python
import clip_feature_extractor
import numpy as np

import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

%load_ext autoreload
%autoreload 2
```

```python
X_train, y_train, X_test, y_test = clip_feature_extractor.get_CIFAR100_features()
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'objective': 'multi:softmax',  # For classification
    'num_class': len(set(y_train)), # Number of classes
    'eval_metric': 'mlogloss',     # Evaluation metric
    'max_depth': 16,               # Increased maximum depth of a tree
    'eta': 0.1,                    # Lower learning rate for better convergence
    'subsample': 0.8,              # Subsample ratio to prevent overfitting
    'colsample_bytree': 0.8,       # Fraction of features to be randomly sampled for each tree
    'device': 'cuda',
    'lambda': 1,                   # L2 regularization term
    'alpha': 0.5                   # L1 regularization term
}

num_rounds = 500  
bst = xgb.train(params, dtrain, num_rounds)

preds = bst.predict(dtest)

accuracy = accuracy_score(y_test, preds)
print(f'Accuracy: {accuracy:.4f}')
print(classification_report(y_test, preds))

conf_matrix = confusion_matrix(y_test, preds)
print("Confusion Matrix:")
print(conf_matrix)
```
