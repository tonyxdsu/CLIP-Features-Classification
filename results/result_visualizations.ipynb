```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ['Logistic Regression', 'SVM', 'KNN', 'Random Forest', 'XGBoost', 'Neural Network']
metrics = ['CIFAR10', 'CIFAR100', 'MNIST']
values = [
    [95.12, 80.22, 98.89],  
    [42, 42, 42],
    [92.77, 71.53, 96.79],  
    [42, 42, 42],  
    [42, 42, 42],  
    [95.15, 80.55, 99.05],
]

values = np.array(values).T

x = np.arange(len(models))  # Group locations
width = 0.275  # Bar width

fig, ax = plt.subplots(figsize=(12, 6))
for i, metric in enumerate(metrics):
    bars = ax.bar(x + i * width, values[i], width, label=metric)
    # Add exact values on top of each bar
    for bar in bars:
        yval = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2, 
            yval + 0.5, 
            f"{yval:.2f}", 
            ha='center', 
            va='bottom', 
            fontsize=10
        )

# Formatting
ax.set_xlabel('Models')
ax.set_ylabel('Test Accuracy %')
ax.set_title('Test Accuracies of Models Across Datasets')
ax.set_xticks(x + width * (len(metrics) - 1) / 2)
ax.set_xticklabels(models)
ax.legend(loc='lower left', bbox_to_anchor=(0.01, -0.15), ncol=3, fontsize=10)

plt.tight_layout()
plt.show()
```
